## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, how="inner", on="Mouse ID")

# Display the data table for preview
df
```

```python
# Checking the number of mice.
mouse_count = len(df["Mouse ID"].value_counts())
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = df[df.duplicated(['Mouse ID', 'Timepoint'])]
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID.

# Sort data by setting Mouse ID as index
duplicates_df = df.set_index('Mouse ID')
duplicates_df.loc['g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = duplicates_df.drop(['g989'])
clean_df
```

```python
# Reset Index to attain Mouse ID column for statiscal use
clean_df2 = clean_df.reset_index()

# Checking the number of mice in the clean DataFrame.
len(clean_df2['Mouse ID'].value_counts())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

tumor = clean_df2.groupby(['Drug Regimen'])

# Statistics
tumor_mean = tumor['Tumor Volume (mm3)'].mean()
tumor_med = tumor['Tumor Volume (mm3)'].median()
tumor_var = tumor['Tumor Volume (mm3)'].var()
tumor_std = tumor['Tumor Volume (mm3)'].std()
tumor_sem = tumor['Tumor Volume (mm3)'].sem()

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

# Data Frame
summary_df = pd.DataFrame({
                        "Mean": tumor_mean,
                        "Median" : tumor_med,
                        "Variance" : tumor_var,
                        "Standard Deviation" : tumor_std,
                        "SEM" : tumor_sem
                        })
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_tm = clean_df2.groupby(['Mouse ID'])['Timepoint'].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

# Convert the panda series into a DataFrame
last_tm_df = pd.DataFrame(last_tm)

# Use on to avoid duplicates and 'right' to only include max timepoints
merged_df = pd.merge(clean_df2, last_tm_df, on=('Mouse ID', 'Timepoint'), how = 'right')
merged_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor volumes
tumor_cap = []
tumor_ram = []
tumor_inf = []
tumor_cef = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Locate the rows which contain mice on each drug and get the tumor volumes
for index, row in merged_df.iterrows():
    
    if row["Drug Regimen"] == treatments[0]:
        tumor_cap.append(row['Tumor Volume (mm3)'])
        
    if row["Drug Regimen"] == treatments[1]:
        tumor_ram.append(row['Tumor Volume (mm3)'])
        
    if row["Drug Regimen"] == treatments[2]:
        tumor_inf.append(row['Tumor Volume (mm3)'])
        
    if row["Drug Regimen"] == treatments[3]:
        tumor_cef.append(row['Tumor Volume (mm3)'])
    
# add subset 
drug_df = pd.DataFrame({treatments[0]: tumor_cap,
                       treatments[1]: tumor_ram,
                       treatments[2]: tumor_inf,
                       treatments[3]: tumor_cef
                       })
    
#Determine outliers using upper and lower bounds

# Create empty lists for outlier table
lowq_2 = []
uppq_2 = []
iqr_2 = []
lowb_2 = []
uppb_2 = []
max_2 = []
min_2 = []

# Loop through each drug
for drug in drug_df:
    quartiles = drug_df[drug].quantile([.25,.5,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    maximum = drug_df[drug].max()
    minimum = drug_df[drug].min()
    
    # Appends the values into a list
    lowq_2.append(lowerq)
    uppq_2.append(upperq)
    iqr_2.append(iqr)
    lowb_2.append(lower_bound)
    uppb_2.append(upper_bound)
    max_2.append(maximum)
    min_2.append(minimum)
    
outliers = pd.DataFrame({'Drug': treatments,
                         'Lower Quartile':lowq_2,
                         'Upper Quartile':uppq_2,
                         'IQR':iqr_2,
                         'Lower Bound': lowb_2,
                         'Upper Bound': uppb_2,
                         'Max': max_2,
                         'Min': min_2
                        })
outliers_df = outliers.set_index('Drug')
outliers_df
```

```python
# Determining outliers using loop

for index, row in outliers_df.iterrows():
    
    if row['Lower Bound'] > row['Min']:
        print(f'There is an outlier in {index}, the minimum value is out of bounds.')
    
    if row['Upper Bound'] < row['Max']:
        print(f'There is an outlier in {index}, the maximum value is out of bounds.')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Convert dataframe into list for plotting
boxplot_list = [drug_df[treatments[0]],
                drug_df[treatments[1]],
                drug_df[treatments[2]],
                drug_df[treatments[3]]
               ]

fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.set_title('Final Tumor Volume by Drug Regimen')
ax1.set_xlabel('Drug Regimen')
ax1.set_ylabel('Tumor Vol (mm3)')

ax1.boxplot(boxplot_list, )
plt.xticks([1,2,3,4], treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

capomulin_df = clean_df2.loc[clean_df2['Mouse ID'] == 'b742', :]

x_axis = capomulin_df['Timepoint']
y_axis = capomulin_df["Tumor Volume (mm3)"]


plt.plot(x_axis, y_axis, linewidth=2,) 
plt.title('Capomulin treatmeant of mouse b742')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')

plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

# Isolate data for Capomulin regimen
cap_df = clean_df2.loc[clean_df2['Drug Regimen'] == "Capomulin", :]

# Get averages
cap_avg = cap_df.groupby(['Mouse ID']).mean()

x_values = cap_avg['Weight (g)']
y_values = cap_avg['Tumor Volume (mm3)']

plt.scatter(x_values, y_values, marker="o", facecolors="red", edgecolors="black",
            s=y_values*2)
plt.title('Average Tumour Volume per Mice Weight (Capomulin Regimen)')
plt.xlabel('Weight of Mouse (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(x_values, y_values)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}")
```

```python
# Linear Regression Model
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values,y_values, marker="o", facecolors="blue", edgecolors="black", alpha=0.75)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,37),fontsize=15,color="red")
plt.title('Average Tumour Volume per Mice Weight (Capomulin Regimen)')
plt.xlabel('Weight of Mouse (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
